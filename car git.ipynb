```python
import matplotlib.pyplot as plt
import seaborn as sns
```

# Show the histogram os horsepower

```python
sns.histplot(car['Horsepower'],kde=True)
plt.show()
```

# show the scatterplot of weight vs horsepower

```python
sns.scatterplot( x = 'Weight' , y = 'Horsepower' , data=car)
plt.show()
```

# WEIGHT DISTRIBUTION BY TYPE

```python
sns.boxplot(x='Type' , y='Weight' , data=car)
plt.title('Weight Distribution by Type')
plt.show()
```

# most_common_type_by_drivetrain

```python
most_common_type_by_drivetrain = car.groupby('DriveTrain')['Type'].agg(pd.Series.mode)
print(most_common_type_by_drivetrain)
```

# Engine size vs Horsepower

```python
sns.scatterplot(x='EngineSize' , y='Horsepower' , data=car)
plt.title('Engine size vs Horsepower')
plt.show()
```

```python
sns.pairplot(car[['EngineSize', 'Horsepower', 'Weight', 'MSRP']])
plt.show()
```
